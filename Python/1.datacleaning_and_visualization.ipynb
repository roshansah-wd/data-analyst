# Data Cleaning

```python
# How to check Null value of single dataset
dataset["artist_genres"].isnull().sum()
```

```python
# data cleaning
dataset["artist_genres"].fillna(dataset["artist_genres"].mode()[0])
```

```python
dataset.head(3)
```

```python
dataset["artist_genres"]=(dataset["artist_genres"].replace(r"^\s*$",np.nan,regex=True).fillna("Unknown"))
```

# Data Visulization

```python
counts=dataset["artist_genres"].value_counts()
plt.bar(counts.index,counts.values)
plt.xlabel("Generes of Artist")
plt.ylabel("Counts")
plt.title("Distribution of Artist Generes")
plt.xticks(rotation=45)
plt.show()
```

```python
dataset.head(2)
```

```python
top=dataset["artist_genres"].value_counts().head(20)
plt.figure(figsize=(8,6))
plt.barh(top.index,top.values)
plt.xlabel("Count",size=15,color='green',family='serif')
plt.ylabel("Top 20 Artist Genres",size=15,color='green',family='serif')
plt.title("Visualization of Art_Genres",size=25,color='red',family="serif")
plt.show()
```

```python
name=dataset["track_name"].value_counts().head(20)
plt.figure(figsize=(8,6))
plt.barh(name.index,name.values)
plt.xlabel('Count')
plt.ylabel('Track Name')
plt.title('Visualization of Track Name')
plt.show()
```
