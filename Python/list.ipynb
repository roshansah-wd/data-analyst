## Iteration with while loops

```python
result = 1
i = 1

while i<=1000:
    result = result * i
    i = i + 1
print('The factorial of 100 is: {}'.format(result))
```

```python
%%time
result = 1
i = 1

while i<=1000:
    result = result * i
    i = i + 1
print('The factorial of 100 is :{}'.format(result))
```

```python
line = '*'
max_length = 10

while len(line) <= max_length:
    print(line)
    line +="*"
    
while len(line) > 0:
    print(line)
    line = line[:-1]
```

```python
line = '*'
max_length = 10

while len(line) > 0:
    print(line)
    line = line[:-1]
while len(line) <= max_length:
    print(line)
    line +="*"
    
```

Break Statement

```python
i = 1
result = 1

while i <= 100:
    result = result * i
    if i == 42:
        print('Magic number 42 reached! stopping execution..')
        break
    i+=1
print('i:',i)
print('result:',result)
```

```python
i = 1
result = 1

while i<20:
    i+=1
    if i % 2 == 0:
        print('Skipping{}'.format(i))
        continue
    print('Multiplying with {}'.format(i))
    result = result * i
    
print('i:',i)
print('result:',result)
```

### Iteration with for loops
A for loops is used for iteration or looping over sequences i.e. list tuples, dictionaries, strings and ranges

```python
days = ['Monday','Tuesday','Wednesday','Thrusday','Friday']

for day in days:
    print(day)
```

```python
# Looping over a string
for char in 'Monday':
    print(char)
```

```python
# Looping over a tuple
for fruit in ['Apple','Banana','Guava']:
    print("Here's a fruit:",fruit)
```

```python
# Looping over a dictionary
person = {
    'name':'Roshan sah',
    'sex':'Male',
    'age':20,
    'marriage':True
}
for key in person:
    print("Key:",key,",","Value:",person[key])
```
