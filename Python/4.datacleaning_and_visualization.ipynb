```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Data Cleaning

```python
dataset["Actual Price"]=dataset["Actual Price"].fillna(dataset["Actual Price"].mode()[0])
```

```python
dataset["Actual Price"].isnull().sum()
```

```python
dataset["Saving"]=dataset["Saving"].fillna(dataset["Saving"].mode()[0])
```

```python
dataset["Saving"].isnull().sum()
```

```python
dataset["Rating"]=dataset["Rating"].fillna(dataset["Rating"].mode()[0])
```

```python
dataset["Rating"].isnull().sum()
```

```python
dataset["Reviews"]=dataset["Reviews"].fillna(dataset["Reviews"].mode()[0])
```

```python
dataset["Core"]=(dataset["Core"].replace(r"^\s*$",np.nan,regex=True).fillna("Unknown"))
```

```python
dataset["Core"].head(100)
```

```python
dataset["SSD"]=dataset["SSD"].replace(r"^\s*$",np.nan,regex=True).fillna("Unknown")
```

```python
dataset["SSD"].isnull().sum()
```

```python
dataset.isnull().sum()
```

```python
dataset.head(2)
```

```python
counts=dataset["Brand"].value_counts().head(10)
plt.barh(counts.index,counts.values)
dict1={'family':"serif","size":20,"color":"red"}
dict2={'family':"serif","size":15,"color":"purple"}
plt.xlabel("Pieces",dict2)
plt.ylabel("Brands Name",dict2)
plt.title("No of different Brands of Laptop",dict1)
plt.show()
```

### By visualizing two columns of one dataset

 Key points

Numerical vs Numerical: Scatter, line, hexbin

Categorical vs Categorical: Countplot, stacked bar

Categorical vs Numerical: Boxplot, Violinplot
