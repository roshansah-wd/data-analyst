### Marker Color
markeredgecolor or the shorter mec to set the color of the edge of the markers:

```python
# Set tje EDGE color to red:
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=20,mec='m')
plt.show()
```

### Markerfacecolor or the shorter mfc to set the color insider edge of the markers:


```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=20,mfc='r')
plt.show()
```

```python
# Mark each point with the color named "hotpink":
plt.plot(ypoints,marker='o',ms=15,mec='hotpink',mfc='hotpink')
```

### Creating Scatter Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([5,7,8,7,2,17,9,4,21,1,3])
y=np.array([99,89,67,102,45,32,78,56,45,55,56])

font1={'family':'serif'}
font2={'family':'serif'}
plt.title("Chicken Pox",color='purple',size='10',fontdict=font1)
plt.xlabel("Meter",color='red',size='15',fontdict=font2)
plt.ylabel("gram",color='green',size='15',fontdict=font2)
plt.suptitle("America",fontdict=font1,size='30')
plt.scatter(x,y,color='red')
plt.grid()
plt.show()
```

### ColorMap

```python
# Create a color array, and specify a colormap in the scatter plot.
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x,y,c=colors,cmap='viridis')
plt.show()
```

```python
# Create a color array, and specify a colormap in the scatter plot.
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x,y,c=colors,cmap='BuPu')
plt.colorbar()
plt.show()
```

## Size

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

### Set your own size for the markers:


```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes,alpha=0.1)

plt.show()

```

### Combine Color Size and Alpha

```python
x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100,size=(100))
sizes=10*np.random.randint(100,size=(100))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='nipy_spectral')
plt.title("ColorBar",size=30,color='red')
plt.xlabel("x-axis",size=20,color='magenta',family='serif')
plt.ylabel('y-axis',size=20,color='brown',family='serif')
plt.colorbar()
plt.grid(axis='o')
plt.show()

```

### Start Angle

```python
y = np.array([35,25,25,15])
mylabels=["Apple","Bananas","Cherries","Dates"]

plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

```python
y = np.array([35,25,25,15])
mylabels=["Apple","Bananas","Cherries","Dates"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,startangle=90,explode=myexplode)
plt.show()
```

##  Shadow

```python
y = np.array([35,25,25,15])
mylabels=["Apple","Bananas","Cherries","Dates"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,startangle=90,explode=myexplode,shadow=True)
plt.show()
```

```python
y=np.array([10,30,20,40])
mylabels=["Bolero","Lamborgini","Bucatti","Rolls Rolyce"]
mycolors=["Black","Red","Grey","Hotpink"]

plt.pie(y,labels=mylabels,colors=mycolors)
plt.show()
```

### Legend
To add a list of explanation for each wedge, use the legend()

```python
y=np.array([10,30,20,40])
mylabels=["Bolero","Lamborgini","Bucatti","Rolls Rolyce"]
mycolors=["Black","Red","Grey","Hotpink"]

plt.pie(y,labels=mylabels,colors=mycolors)
plt.legend(title="Cars")
plt.show()
```
