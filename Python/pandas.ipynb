How to handle time series data with ease

```python
air_quality["datetime"]=pd.to_datetime(air_quality["datetime"])
air_quality["datetime"]
```

```python
pd.read_csv("../data/air_quality_no2_long.csv",parse_dates=["datetime"])
```

```python
# What is the start and end date of the time series data set we are working with?
air_quality["datetime"].min(),air_quality["datetime"].max()
```

```python
air_quality["datetime"].max()-air_quality["datetime"].min()
```

```python
# To add a new column to the dataframe containing only the month of the measurement
air_quality["month"]=air_quality["datetime"].dt.month
air_quality.head()
```

```python
# what is the average NO2 concentration for each day of the week for each of the measurement locations?
air_quality.groupby(
    [air_quality["datetime"].dt.weekday,"location"]  
)["value"].mean()
```

```python
fig,axs=plt.subplots(figsize=(12,4))
air_quality.groupby(air_quality["datetime"].dt.hour)["value"].mean().plot(
    kind='bar',rot=0,ax=axs
)
plt.xlabel("Hour of the day")
plt.ylabel("$NO_2(ug/m^3)$")
```

### Resample a time series  to another frequency

```python
# Aggregate the current hourly time series values to the monthly maximum value in each of the stations.
monthly_max=no_2.resample("Me").max()
monthly_max
```

```python
# Start by creating a series with 9 one minute timestamps
index=pd.date_range('1/1/2000',periods=9,freq='min')
series=pd.Series(range(9),index=index)
series
```

```python
series.resample('3min').sum()
```

```python
#  To include the vlue close the right side of the bin interal, as shown below.
series.resample('3min',label='right',closed='right').sum()
```

```python
# Unsample the series into 30 second bins
series.resample('30s').asfreq()[0:5]
```

```python
# Unsample the series into a 30 second bins and fill the NaN values using the ffill method.
series.resample('30s').ffill()[0:5]
```

```python
# Unsample the series into a 30 second bins and fill the NaN values using the ffill method.
series.resample('30s').bfill()[0:5]
```

```python
import numpy as np
```

```python
 #Pass a custom funnction via apply
def custom_resampler(arraylike):
    return np.sum(arraylike)+5
```

```python
series.resample('3min').apply(custom_resampler)
```

```python
# Make a plot of the daily mean No2 value in each of the stations.
no_2.resample("D").mean().plot(style="o",figsize=(10,5))
```
